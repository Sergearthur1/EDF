```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt


plt.rcParams['figure.figsize'] = 15, 7
```

```python
def parser1(x):
    x=str(x)[:19]
    return pd.datetime.strptime(x, '%Y-%m-%dT%H:%M:%S')

def parser2(x):
    x=str(x)
    return pd.datetime.strptime(x, '%d/%m/%y %Hh%M')
```

```python
df_temp = pd.read_csv('elec_train.csv', sep=';', parse_dates=[0], date_parser=parser1)
meteo1 = pd.read_csv('meteo_train.csv', sep=';', parse_dates=[0], date_parser=parser2)
meteo2 = pd.read_csv('meteo_prev.csv', sep=';', parse_dates=[0], date_parser=parser2)
meteo = pd.concat([meteo1,meteo2])

meteo.rename(columns={'Date UTC':'date'},inplace=True)
meteo.head()
```

```python
df_temp['date'] = df_temp['date'].dt.round('1h')
df_temp.head()
```

```python
df = pd.merge(df_temp, meteo, on=['date'],how='outer',sort=True)
df.rename(columns={'T¬∞ (C)': 'Temperature (C)',
                       'P (hPa)': 'Pression (hPa)',
                      'HR (%)': 'Humidité Relative'},inplace=True)
df = df.interpolate()
df = df.round(3)
df
```

On calcul la température ressentie ou le refroidissement éolien. Jugé utile car beaucoup de vent dans cette région.

```python
df['Temperature ressentie'] = round(13.2+0.6215*df['Temperature (C)']+(0.3965*df['Temperature (C)']-11.37)*df['Vt. raf. (km/h)']**(0.16),3)
df.head()
```
